```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv("../StudentsPerformance.csv")
```

```python
students_performance
```

```python
students_performance.head()
students_performance = students_performance\
                                    .rename(columns=
                                            {'parental level of education': 'parental_level_of_education',
                                            'test preparation course': 'test_preparation_course',
                                            'math score': 'math_score',
                                            'reading score': 'reading_score',
                                            'writing score': 'writing_score'})
```

```python
students_performance.groupby("gender", as_index=False)\
    .aggregate({"math_score": "mean", "reading_score": "mean"})\
    .rename(columns={'math_score': "mean_math_score", 'reading_score': 'mean_reading_score'})
```

```python
mean_scores = students_performance.groupby(["gender", 'race/ethnicity'])\
    .aggregate({"math_score": "mean", "reading_score": "mean"})\
    .rename(columns={'math_score': "mean_math_score", 'reading_score': 'mean_reading_score'})
```

```python
mean_scores.loc[("female", 'group B')]
```

```python
students_performance.groupby(["gender", 'race/ethnicity'])\
    .reading_score.nunique()
```

```python
students_performance.sort_values(["gender", "math_score"], ascending=False)\
    .groupby('gender').head()
```

```python
students_performance["total_score"] = students_performance.math_score + students_performance.reading_score + students_performance.writing_score
```

```python
students_performance.math_score + students_performance.reading_score + students_performance.writing_score
```

```python
students_performance
```

```python
students_performance = students_performance.assign(total_score_log =\
                                                   np.log(students_performance.total_score))
```

```python
students_performance
```

```python
students_performance.drop(["total_score"], axis=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
students_performance = pd.read_csv("../StudentsPerformance.csv")
students_performance.head()
students_performance = students_performance\
                                    .rename(columns=
                                            {'parental level of education': 'parental_level_of_education',
                                            'test preparation course': 'test_preparation_course',
                                            'math score': 'math_score',
                                            'reading score': 'reading_score',
                                            'writing score': 'writing_score'})
```

```python
students_performance.plot.scatter(x="math_score", y="reading_score")
```

```python
ax = sns.lmplot(x="math_score", y="reading_score", hue="gender",\
           data=students_performance, fit_reg=False)
ax.set_xlabels("Math_score")
ax.set_ylabels("Reading_score")
```
