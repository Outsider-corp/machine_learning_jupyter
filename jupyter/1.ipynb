```python
dota = pd.read_csv("../dota_hero_stats.csv")
```

```python
dota.roles.str.split(",")
```

```python
d = pd.DataFrame(columns=["roles"])
for i in dota.roles.str.split(","):
#     d.append([j.replace('[',"").replace(']',"") for j in i])
    for j in i:
        d = d.append({'roles':j.replace('[',"").replace(']',"").strip()}, ignore_index=True)
d.groupby("roles").aggregate({"roles":"count"})
```

```python
dd = pd.DataFrame(columns=["roles"])
for i in dota.roles:
    dd = dd.append({'roles':len(i.split(","))}, ignore_index=True)
dd.groupby('roles').aggregate({'roles':'count'})
```
